https://gist.github.com/rickiepark/1e89fe2a9d4ad92bc9f073163c9a37a7 빙어 14마리의 데이터  
이제 도미와 빙어 데이터를 함께 산점도로 그리기 !!  



```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
len(smelt_length)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

사이킷런을 쓰려면 2차원 리스트가 필요!!

```python
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
print(fish_data)
```

이제 생선 49개의 길이와 무게 데이터셋을 준비했으니 정답 데이터를 만들어야함! -> 원하는 거: 이게 도미인지 빙어인지?  
도미:1 빙어:0 으로 하자!!

```python
fish_target = [1]*35 + [0]*14 
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target)
```

이 값을 **정확도**라고 부르고 정확도가 100%인 상태!  
이제 k-최근접 이웃 알고리즘을 더 자세히 알아보기 위해 새로운 데이터가 도미인지 빙어인지 알아보자!!

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30,600, color = 'green', marker='^')  #새로운 물고기!!
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn.predict([[30, 600]])
```
